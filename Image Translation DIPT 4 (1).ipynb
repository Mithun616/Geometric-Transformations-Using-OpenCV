# **Image Translation**

Write a Python program using OpenCV to perform image translation on the **"Given image"**.

The program should:

1. Import `cv2`, `numpy`, and `matplotlib.pyplot`.
2. Read the image **"given** in color mode.
3. Create a **translation matrix** that moves the image  
   **50 pixels to the right** and **80 pixels down**.
4. Apply the translation
5. Display the following using a **1 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Translated Image**


```python
# Import 
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
img = cv2.imread("baseball.jpg")
```

```python
# Display the image using Matplotlib
plt.imshow(img[:,:,::-1])
plt.title("Original Image")
plt.axis("off")
```

```python
# Translation 
M = np.float32([[1, 0, 50],
                [0, 1, 80]])
translated = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Display original and translated images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img[:,:,::-1])
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(translated[:,:,::-1])
plt.title("Translated Image")
plt.axis("off")

plt.show()
```

# **Image Shearing**

Write a Python program using OpenCV to perform image shearing on the **given image**.

The program should:

1. Create a **shearing transformation matrix** to apply:
   - **Horizontal shearing**  
   - **Vertical shearing**
2. Apply the shearing transformations.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Sheared Image**  
   - **Vertically Sheared Image**


```python
# Get image height and width
h, w = img.shape[:2]
```

```python
# Horizontal shearing matrix
M1 = np.float32([[1, 0.5, 0],
                 [0, 1, 0]])

# Vertical shearing matrix
M2 = np.float32([[1, 0, 0],
                 [0.5, 1, 0]])
```

```python
# Apply shearing
h_shear = cv2.warpAffine(img, M1, (w, h))
v_shear = cv2.warpAffine(img, M2, (w, h))
```

```python
# Display images
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img[:,:,::-1])
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(h_shear[:,:,::-1])
plt.title("Horizontal Shear")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(v_shear[:,:,::-1])
plt.title("Vertical Shear")
plt.axis("off")

plt.show()
```

# **Image Rotation**

Write a Python program using OpenCV to perform **image rotation** on the **given image**.

The program should:

1. Rotate the image by:  
   - **45 degrees**  
   - **90 degrees**
2. Use `cv2.getRotationMatrix2D` and `cv2.warpAffine`.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **45° Rotated Image**  
   - **90° Rotated Image**


```python
# Get image height and width
h, w = img.shape[:2]
```

```python
# Find center of image
center = (w//2, h//2)
```

```python
# Rotation matrix for 45 degree
M1 = cv2.getRotationMatrix2D(center, 45, 1)
```

```python
# Rotation matrix for 90 degree
M2 = cv2.getRotationMatrix2D(center, 90, 1)
```

```python
# Rotate images
rotated_45 = cv2.warpAffine(img, M1, (w, h))
rotated_90 = cv2.warpAffine(img, M2, (w, h))
```

```python
# Display images
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img[:,:,::-1])
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(rotated_45[:,:,::-1])
plt.title("45° Rotated Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(rotated_90[:,:,::-1])
plt.title("90° Rotated Image")
plt.axis("off")

plt.show()
```
